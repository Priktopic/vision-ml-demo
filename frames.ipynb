```python
os.environ["GOOGLE_APPLICATION_CREDENTIALS"]="/home/priyanka/Downloads/projectcup-751f24d994bc.json"

from google.cloud import storage

storage_client=storage.Client()
bucket = storage_client.get_bucket(bucket_name)

def upload(src_name, dst_name):
    blob = bucket.blob(dst_name)
    blob.upload_from_filename(src_name)
```

```python
# Read the video from specified path 
cam = cv2.VideoCapture("demo-video.mp4") 

try: 
    # creating a folder 
    if not os.path.exists('data'): 
        os.makedirs('data') blob.upload_from_filename("cup_data.csv")

# if not created then raise error 
except OSError: 
    print ('Error: Creating directory of data') 

# frame 
currentframe = 0

data_array = []

while(True): 
    # reading from frame 
    ret,frame = cam.read() 

    if ret: 
        # if video is still left continue creating images 
        object_name = 'data/frame' + str(currentframe) + '.jpg'
        #print ('Creating...' + object_name) 

        # writing the extracted images 
        cv2.imwrite(object_name, frame) 
        
        #upload to google storage
        upload(src_name = object_name, dst_name = object_name)
        
        gcs_path = 'gs://' + bucket_name + "/" + object_name 
        data_array.append((gcs_path,"cup"))
        
        # increasing counter so that it will 
        # show how many frames are created 
        currentframe += 1
    else: 
        break

# Release all space and windows once done 
cam.release() 
cv2.destroyAllWindows() 
```

```python
dataframe=pd.DataFrame(data_array)
dataframe
```

```python
dataframe.to_csv('cup_data.csv', index=False, header=False)
upload("cup_data.csv", "cup_data")
```
